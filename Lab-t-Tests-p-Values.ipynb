```python
# Importing Libs
from scipy.stats import ttest_1samp, ttest_rel, ttest_ind
from scipy.stats import f_oneway
from scipy.stats import norm
from statistics import math

import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Part 0 - Exercise 1
df_machine = pd.read_csv(r"C:\Users\joaoa\Desktop\Ironhack\Labs\lab-t-tests-p-values\files_for_lab\ttest_machine.txt", sep = " ")
df_machine
```

```python
from scipy.stats import ttest_ind
data1 = df_machine["New_machine"]
data2 = df_machine["Old_machine"]

stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p/2))
if p > 0.05:
    print('There is a 95% level of confidence that the new machine is better than the old one')
else:
    print('There is a 95% level of confidence that the new machine is better than the old one')
```

```python
# Part 0 - Exercise 2
df_pokemon = pd.read_csv(r"C:\Users\joaoa\Desktop\Ironhack\Labs\lab-t-tests-p-values\files_for_lab\pokemon.txt")
df_pokemon
```

```python
# Example of the Paired Student's t-test

data1 = df_pokemon["Attack"]
data2 = df_pokemon["Defense"]
stat, p = ttest_rel(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('There is a 95% level of confidence that the defense and attack scores are equal')
else:
    print('There is a 95% level of confidence that the defense and attack scores are different')
```

```python
# Part 2
df_ANOVA = df_pokemon.iloc[:,2:5]
df_ANOVA = df_ANOVA.drop("Type 2", axis = 1)
df_ANOVA.columns = ["Type_1","Total"]
df_ANOVA
```

```python
model = ols("Total ~ C(Type_1)", data = df_ANOVA).fit()

table = sm.stats.anova_lm(model)
table

# Since p_value < 0.05, we reject de null hypothesis
# It means that we are 95% confident that the means are not the same
```
