```python
pokemon = pd.read_csv(r"C:\Users\pedro\Desktop\Ironhack\pokemon.txt")
pokemon
```

```python
pokemon.rename(columns={'Type 1': 'Type_1'}, inplace=True)
```

```python
#dependent samples t-test
#data provides sufficient evidence that there is a statistically significant difference 
#between attack and defense scores (p-value<0.01)
ttest_rel(pokemon['Attack'],pokemon['Defense'])
```

```python
#ANOVA test
#data provides sufficient evidence that not all population means are equal (p-value<0.01)
model = ols("Total ~ Type_1", data = pokemon).fit()

table = sm.stats.anova_lm(model)
table
```
