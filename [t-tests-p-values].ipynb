```python
# Libaries

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import ttest_ind, ttest_rel
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import f_oneway
```

```python
'''
Instructions
# 1.
One tailed t-test - In a packing plant, a machine packs cartons with jars. It is supposed that a new machine will pack 
faster on the average than the machine currently used. To test that hypothesis, the times it takes each machine to pack 
ten cartons are recorded. The results, in seconds, are shown in the tables in the file files_for_lab/ttest_machine.xlsx.. 
Assume that there is sufficient evidence to conduct the t test, does the data provide sufficient evidence to show if one 
machine is better than the other?
'''

# a = 0.05
```

```python
df = pd.read_csv(r"C:\Users\ljant\Desktop\Ironhack\15_Revision_and_Hypothesis-Testing\lab-t-tests-p-values\files_for_lab\ttest_machine.txt", sep='\s+', header=0)
df
```

```python
t_statistic, p_value = ttest_ind(df['New_machine'], df['Old_machine'])
print(t_statistic, p_value)
```

```python
'''
Answer:
Since p_value is < than a, I'm rejecting H0 and assuming with 95 % certainty that the new machine is significantly better
/more efficient than the older version. 
'''
```

```python
'''
# 2.
Matched Pairs Test - In this challenge we will compare dependent samples of data describing our Pokemon 
(file files_for_lab/pokemon.csv). Our goal is to see whether there is a significant difference between each Pokemon's 
defense and attack scores. Our hypothesis is that the defense and attack scores are equal. Compare the two columns to see 
if there is a statistically significant difference between them and comment your result.
'''
# a = 0.05
```

```python
df = pd.read_csv(r"C:\Users\ljant\Desktop\Ironhack\15_Revision_and_Hypothesis-Testing\lab-t-tests-p-values\files_for_lab\pokemon.txt")
df
```

```python
t_statistic, p_value = ttest_rel(df['Attack'], df['Defense'])
print(t_statistic, p_value)
```

```python
'''
Answer:
Since p_value is > than a, I'm rejecting H1 and assuming with 95 % certainty that there are no significant differences
between the values for Attack and Defense per Pokemon.
'''
```
