```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')



# Importing Libs
from scipy.stats import ttest_1samp, ttest_rel, ttest_ind
from scipy.stats import f_oneway
from scipy.stats import norm
from statistics import math
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# One tailed t-test

# importing the machines packing data 

packing = pd.read_csv('/Users/giulianamiranda/Documents/Labs/lab-t-tests-p-values/files_for_lab/ttest_machine.txt', sep=' ')
packing
```

```python
# Defining the null hypothesis
# H0 is the mean of both machines is equal

# We are stting the alpha (significance level) to 5%, the industry standard

alpha = 0.05

# conducting the T thest

t_statistic, p_value = ttest_ind(packing['New_machine'], packing['Old_machine'])

#getting the result
print('T statistic is:', t_statistic,'P value is:', p_value)
```

```python
# Since the P value (0.0032) is lower than the alpha (5%), we are 95% sure that the means of both machines are different
```
