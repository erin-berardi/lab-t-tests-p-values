```python
# Part 1

# Null hypothesis - Alternate hypothesis - Level of significance - Test statistic - P-value - F table


# first checking the unique values for all feature of the dataframe

unique_values_list = []  

for column in pokemon_data:
    unique_values = pokemon_data[column].unique()
    unique_values_list.append((column, unique_values))

# Now unique_values_list contains tuples with column name and unique values
# You can access the unique values for each column as needed
for column_name, unique_values in unique_values_list:
    print(f'Unique values in {column_name}:')
    print(unique_values)
    print()
```

```python
# now I will create a list with for the total for each different type combined between 1 and 2 

column1 = pokemon_data["type_1"]

column2 = pokemon_data["type_2"]

# grouping by column type and getteing the mean for each total 
Type2_lst = pokemon_data.groupby([column1,column2])['total'].mean().reset_index()

# doing the same but with the totals

Type_sum_lst = pokemon_data.groupby([column1,column2])['total'].sum().reset_index()

# rounding result of mean of totals for the two type combined 
Type2_lst["total"] = Type2_lst["total"].round(1)

# pokemon_totals = Type2_lst
pokemon_totals = Type2_lst 
pokemon_totals
```

```python
# will combine type 1 and type 2 column to get only one feature 

pokemon_totals['Type_combined'] = pokemon_totals['type_1'] + '_' + pokemon_totals['type_2']

pokemon_totals['Type_combined']

# to be able to use the model Anova
```

```python
# tried to use Anova model on the list created but it does not work and give bad result 

# will run Anova test on  pokemon_totals Data Frame, and type combined column 
model = ols( 'total ~ C(Type_combined)', data = pokemon_totals).fit()

table = sm.stats.anova_lm(model)

table
```

```python
#  Third we run ANOVA test on pokemon_data.
 

# State the null hypothesis

# null hypothesis is that all means are equal 

#  State the alternate hypothesis

# alternate hypothesis is that means are different 

# What is the significance level

# significance level is 5% , alpha is 5%

# This is a good balance between sensitivity and specificity.

```

```python
# will run Anova test on type 1 column on data frame pokemen_data
# gives good result 
# P value is really small 
model = ols( 'total ~ C(type_1)', data = pokemon_data).fit()

table = sm.stats.anova_lm(model)

table


# Degrees of Freedom for Model : 17
# Residual Degrees of Freedom : 782
# Total Degrees of Freedom : 799 ( residual + dof model)
```

```python
# will run Anova test on type 2 column on data frame pokemen_data
# same thing the p value is really low number
model = ols( 'total ~ C(type_2)', data = pokemon_data).fit()

table = sm.stats.anova_lm(model)

table
# What are the degrees of freedom of model, error terms ( residual ), and total DoF

# Degrees of Freedom for Model : 17
# Residual Degrees of Freedom : 396.0
# Total Degrees of Freedom : 413 ( residual + dof model)
```

```python
# What conclusions can you draw from the experiment and why?
# Interpret the ANOVA test result. Is the difference significant?



# The p-value for the 'Type 1' and type 2 variable is very small for both model ,
# less than significance level (0.05)
#. When the p-value is < alpha, 
# we reject the null hypothesis that the group means are equal. 

# Yes the difference is significant bcs p value is small , so the total values between groups of 'Type 1' variable are not equal.

```
